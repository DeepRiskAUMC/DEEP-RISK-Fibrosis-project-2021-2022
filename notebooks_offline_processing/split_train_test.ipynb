```python
from sklearn.model_selection import train_test_split
import shutil
import os

path = '/home/empquist/VAE_EM/processed_emidec_dataset'
new_path = '/home/empquist/VAE_EM/emidec_processed_dataset'

path_myo = '/myo_lv/binMask'
path_img = '/whole_image/Training_cropped'


if not os.path.exists(os.path.join(new_path, 'test' + path_myo)):
    os.makedirs(os.path.join(new_path, 'test' + path_myo), exist_ok=True)

if not os.path.exists(os.path.join(new_path, 'test' + path_img)):
    os.makedirs(os.path.join(new_path, 'test' + path_img), exist_ok=True)

if not os.path.exists(os.path.join(new_path, 'train' + path_myo)):
    os.makedirs(os.path.join(new_path, 'train' + path_myo), exist_ok=True)

if not os.path.exists(os.path.join(new_path, 'train' + path_img)):
    os.makedirs(os.path.join(new_path, 'train' + path_img), exist_ok=True)

train_ratio = 0.8

all_patient_ids = os.listdir(path + path_myo)
train_patient_ids, test_patient_ids = train_test_split(all_patient_ids, train_size=0.8, test_size=0.2)


for data_set, name in [(train_patient_ids, 'train'), (test_patient_ids, 'test')]:
    for scan in data_set:
        print(scan)
        nr = '_'.join(scan.split("_")[:2])
        print(nr, name)
        src_patient_folder_myo = os.path.join(path + path_myo, scan)
        dst_patient_folder_myo = os.path.join(new_path, name + path_myo, scan)
        shutil.copy2(src_patient_folder_myo, dst_patient_folder_myo)

        src_patient_folder_img_gt_scar = os.path.join(path + path_img, nr + '_gt_scar.nii.gz')
        dst_patient_folder_img_gt_scar = os.path.join(new_path, name + path_img, nr + '_gt_scar.nii.gz')
        shutil.copy2(src_patient_folder_img_gt_scar, dst_patient_folder_img_gt_scar)

        src_patient_folder_img_gt = os.path.join(path + path_img, nr + '_gt.nii.gz')
        dst_patient_folder_img_gt = os.path.join(new_path, name + path_img, nr + '_gt.nii.gz')
        shutil.copy2(src_patient_folder_img_gt, dst_patient_folder_img_gt)

        src_patient_folder_img = os.path.join(path + path_img, nr + '.nii.gz')
        dst_patient_folder_img = os.path.join(new_path, name + path_img, nr + '.nii.gz')
        shutil.copy2(src_patient_folder_img, dst_patient_folder_img)
```
