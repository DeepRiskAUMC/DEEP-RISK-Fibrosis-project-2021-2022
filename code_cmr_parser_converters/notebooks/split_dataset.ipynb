```python
from sklearn.model_selection import train_test_split
import shutil
import os
```

```python
path = 'L:\\basic\Personal Archive\E\emquist\parsing_MRI\\final_test_seg\LGE_clin_myo_ICM_labeled_filtered_3y'
new_path = 'L:\\basic\Personal Archive\E\emquist\parsing_MRI\\final_test_seg\ICM_LGE_clin_myo_mask_labeled_filtered_3y'

if not os.path.exists(os.path.join(new_path, 'test')):
    os.makedirs(os.path.join(new_path, 'test'), exist_ok=True)

if not os.path.exists(os.path.join(new_path, 'train')):
    os.makedirs(os.path.join(new_path, 'train'), exist_ok=True)

train_ratio = 0.8

all_patient_ids = os.listdir(path)
train_patient_ids, test_patient_ids = train_test_split(all_patient_ids, train_size=0.8, test_size=0.2)

for data_set, name in [(train_patient_ids, 'train'), (test_patient_ids, 'test')]:
    for patient in data_set:
        print(patient, name)
        src_patient_folder = os.path.join(path, patient)
        dst_patient_folder = os.path.join(new_path, name, patient)
        shutil.copytree(src_patient_folder, dst_patient_folder)
```
