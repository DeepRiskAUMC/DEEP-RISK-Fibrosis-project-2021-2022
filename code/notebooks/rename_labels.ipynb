```python
from pathlib import Path
import os

def rename_labels(reference_dir, rename_dir, 
                  current_sequence_identifier_end='_LGE_',
                  new_sequence_identifier_end="-",
                  new_suffix='-aankleuring.nrrd'):
    
    reference_paths = [x for x in reference_dir.glob('*.nrrd')]
    for reference_path in reference_paths:
        # find matching file in output_dir
        sequence_identifier = str(Path(reference_path).stem).split(current_sequence_identifier_end)[0]        
        rename_path = [x for x in rename_dir.glob(f"*{sequence_identifier}*.nrrd")]
        assert len(rename_path) != 0, f'No file found for  {sequence_identifier}' 
        assert len(rename_path) == 1
        rename_path = rename_path[0]
        
        #create new name
        new_sequence_identifier = (Path(reference_path).stem).split(new_sequence_identifier_end)[0]
        new_path = Path(rename_path).parent.joinpath(new_sequence_identifier + new_suffix)
        print(f"{reference_path=}\n {rename_path=}\n {new_path=}\n\n")
        os.rename(rename_path, new_path)
        
        
    return
```
